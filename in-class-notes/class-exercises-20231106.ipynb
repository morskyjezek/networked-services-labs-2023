Activities for APIs in Python.

```python
import requests 
```

```python
endpoint = 'https://www.loc.gov/search'

parameters = {
    'fo' : 'json',
    'q' : 'kittens',
    'fa': 'online-format:image'
}
```

```python
r = requests.get(endpoint, params=parameters)
```

```python
r.url
```

```python
r.status_code
```

```python
search_data = r.json()

print(search_data)
```

```python
r.headers
```

```python
for key in search_data.keys():
    print(key)
```

```python
search_data['results']
```

```python
for result in search_data['results']:
    print(result['url'])
```
