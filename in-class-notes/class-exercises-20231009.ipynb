```python
json_test_string = '''{ "Results": 3, 
"isPrinted" : true, 
"format" : null, 
"books": [
    { "title": "Gone with the Wind", "author":"Mitchell" },
    { "title": "Wuthering Heights", "author": "Bronte" }, 
    { "title": "Adventures of Tom Sawyer", "author": "Twain"}
]}
'''
```

```python
import json 
```

```python
test_json_object = json.loads(json_test_string)

type(test_json_object)
```

```python
test_json_object.keys()
```

```python
test_json_object['Results']
```

```python
test_json_object['books']
```

```python
len(test_json_object['books'])
```

```python
for book in test_json_object['books']:
    print(book['title'])
```

# Getting JSON from the Web

Using requests

```python
import requests 
```

```python
endpoint = 'https://www.loc.gov/free-to-use/statue-of-liberty/'
```

```python
r = requests.get(endpoint)
```

```python
r.status_code
```

```python
r.headers
```

```python
r.text 
```

```python
r.url 
```

```python
parameters = {
    'fo' : 'json'
}
```

```python
r = requests.get(endpoint, params = parameters)
```

```python
r.status_code
```

```python
r.url
```

```python
r2 = requests.get(endpoint, params=parameters)
```

```python
r2.status_code
```

```python
for item in r2.json(): 
    print(item)
```

```python
for key in r2.json().keys():
    print(key)
```
